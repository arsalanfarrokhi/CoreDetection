```python
import cv2
from matplotlib import pyplot as plt
image_path = r'C:\Users\Arsalan\Desktop\CoresResearch\neuralNetwork\poreDetection\PoreGroundTruth\PoreGroundTruthSampleimage'
label_path = r'C:\Users\Arsalan\Desktop\CoresResearch\neuralNetwork\poreDetection\PoreGroundTruth\PoreGroundTruthMarked'
fig = plt.figure(figsize=(30, 30))
rows = 6
columns = 5
for i in range(1, 31):
    img = cv2.imread(image_path + '\\' + f'{i}.bmp')
    fig.add_subplot(rows,columns, i)
    coord = []
    with open(label_path+'\\'+f'{i}.txt', 'r') as file1:
        coord = tuple([int(i) for i in file1.readline().split()])
        #print(coord)
    img = cv2.ellipse(img, coord, (30, 30), 0, 0, 360, (0, 0, 255), 1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f'img: {i}.bmp size: {img.shape}')
```
