```python
import os
import cv2
import numpy 

images_path = r'.\DBII_short'
labels_path = r'.\Labels\\'
dataset_path = r'.\Dataset\\'
idk = r'.\IDK\\'

images_to_go = []
end = 0

for image_path in os.listdir(images_path):
    images_to_go.append(image_path)


for image_path in images_to_go:
    img = images_path + r'\\' + image_path
    
    initial_img = cv2.imread(img, cv2.COLOR_BGR2GRAY)
    img = initial_img.copy()

    cv2.imshow('image', img)
    core_list = []
    imageX, imageY = 0, 0
    #define the events for the
    # mouse_click.

    def mouse_click(event, x, y, flags, param):

        if event == cv2.EVENT_LBUTTONDOWN:
            global imageX, imageY
            cv2.circle(img, (x, y), 14, 255, -1) 
            cv2.imshow('image', img)
            imageX, imageY = x, y
            core_list.append((x, y))
        

    cv2.setMouseCallback('image', mouse_click)
    while(1):
        key = cv2.waitKey(0)
        if(key == ord('w')):
            mask = numpy.zeros(initial_img.shape)
            cv2.circle(mask, (imageX, imageY), 14, 255, -1)
            #cv2.imwrite(labels_path + f'{count}.bmp', mask)
            cv2.imwrite(dataset_path + f'{count}.bmp', initial_img)
            with open(labels_path +  f"{count}.txt", "w") as f:
                for (x, y) in core_list:
                    f.write(f"{x} {y}\n")
            break
        elif(key == ord('i')):
            #cv2.imwrite(labels_path + f'{count}.bmp', mask)
            cv2.imwrite(idk + f'{count}.bmp', initial_img)
            break
        elif(key == ord('c')):
            img = initial_img.copy()

            cv2.imshow('image', img)
            core_list = []
            imageX, imageY = 0, 0            


    cv2.destroyAllWindows()
    #cv2.imwrite(r"C:\Users\arsal\Desktop\NewML\algorithm\PoreGroundTruthSampleimage" + "\\" + str(count) +'.bmp', cv2.imread(images_path + r'\\' + image_path, cv2.COLOR_BGR2GRAY))
```
